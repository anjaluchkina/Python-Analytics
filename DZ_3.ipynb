## Задание 1

Данные находятся в файле kc_house_data.csv.

Считать данные с помощью pandas.

Вывести на экран первые 5 строк.

```python
import pandas as pd

df = pd.read_csv('/content/kc_house_data (1).csv',parse_dates = ['date'])
df.head()
```

### 1.1Создать новый признак price_per_sq_lot, который будет содержать avg = среднюю стоимость за один кв. метр общей площади.

```python
df['price_specific_area']=df['price']/df['sqft_lot']
df.head()
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновации дома и годом постройки дома и сохраните в renv. Если реновации дома не было, то в новом признаке поставьте 0.

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.head()
```

```python
df['delta_renovated'] = df['yr_renovated'].apply(lambda x: x if x > 0 else 0)
df.head()
```

### 1.3 Создайте признаки года продажи, месяца продажи и сохраните в year_ch, month_ch.

```python
df['date'] = pd.to_datetime(df['date'])
df['year_ch'] = df['date'].apply(lambda x: x.date().year)
df['month_ch'] = df['date'].apply(lambda x: x.date().month)
df.sample(5)
```

### 1.4 Удалите признаки date, zipcode, lat, long и сохраните в new.

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```

### Задание 3

### 3.1 среднюю стоимость домов в зависимости от количества спален и сохраните в avg.
Отсортируйте от меньшей стоимости к большей.

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price').apply(lambda x: round(x / 1000,0))
```

### 3.2 Найдите минимальную min, среднюю mean и максимальную max стоимости домов в зависимости от состояния дома и сохраните в price.

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную waterfront и оценкой вида view и сохраните ее в view_waterfront.

```python
df.pivot_table(index='waterfront', columns='view',
               values='condition', aggfunc='count',
               fill_value=0)
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше? Постройте эту таблицу, содержащую в себе информацию о спальнях и этажности, и сохраните ее в bedrooms_floors.

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms'])
```

### 3.5  Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома и сохраните в 'median_price'.

```python
df.pivot_table(index='condition', columns='grade',
               values='price', aggfunc='median',
               fill_value=0)
```
